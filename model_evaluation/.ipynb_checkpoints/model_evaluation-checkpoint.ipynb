```python
# Import modules
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
import sklearn as sk
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('funda-amsterdam.csv')
```

```python
df.head()
```

```python
df1 = df[(df['price'] > 30000) & (df['area'] > 10)]
df_two_variables = df1[['price', 'area']]
```

```python
x = df_two_variables[["area"]]
y = df_two_variables["price"]
```

```python
lm = LinearRegression()
lm.fit(x, y) 
```

```python
df_two_variables['price_p'] = lm.predict(x)
df_two_variables['resid'] = df_two_variables['price'] - df_two_variables['price_p']
```

```python
sns.scatterplot(x='price',y='price_p',data=df_two_variables)

sns.scatterplot(x='price',y='price_p',data=df_two_variables)
plt.xlim(0, 1400000) #This sets the x-axis limits to (0, 5e10 = 5000000000)
plt.ylim(0, 1400000) #Ditto for y-axis. I want both axes to have the same length, so we can compare them
plt.plot([0, 1000000], [0, 1000000], color='red', lw=3) #This draws the straight red line, you can leave this out if you wish
plt.xlabel('Views')
plt.ylabel('Views (predicted)')
plt.show()
```

```python
sns.scatterplot(x='area',y='resid',data=df_two_variables)
```

```python
r2_score(df_two_variables['price'], df_two_variables['price_p'])
```

```python
math.sqrt(mean_squared_error(df_two_variables['price'], df_two_variables['price_p'])) #this gives the model fit (R2) for the Y and predicted Y
```
