```python
# Import modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
# Read csv file student_data.csv and make it readble
df = pd.read_csv("student_data.csv")
```

```python
sns.catplot(x="gender", y="instagram_followers",
            kind="violin", bw=.15, cut=0,
            data=df); # violin plot by gender 
```

```python
male = df[df['gender']=='male'] #this gets us a dataframe with only male
female = df[df['gender']=='female'] #this gets us a dataframe with only female

male_mean = male['instagram_followers'].mean() # calculating the mean of male
female_mean = female['instagram_followers'].mean() # calculating the mean of female
print(f'The mean rating for male like is {male_mean}')
print(f'The mean rating for female like is {female_mean}')
```

```python
ttest_ind(male['instagram_followers'], female['instagram_followers'], nan_policy='omit') # Calculating the t-testing
```
