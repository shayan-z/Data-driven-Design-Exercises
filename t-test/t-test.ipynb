```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("student_data.csv")
```

```python
sns.catplot(x="gender", y="instagram_followers",
            kind="violin", bw=.15, cut=0,
            data=df);
```

```python
male = df[df['gender']=='male'] #this gets us a dataframe with only the paid apps
female = df[df['gender']=='female'] #ditto for the free apps

male_mean = male['instagram_followers'].mean()
female_mean = female['instagram_followers'].mean()
print(f'The mean rating for male like is {male_mean}')
print(f'The mean rating for female like is {female_mean}')
```

```python
ttest_ind(male['instagram_followers'], female['instagram_followers'], nan_policy='omit')
```
