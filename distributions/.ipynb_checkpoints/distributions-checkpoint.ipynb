```python
# Import modules
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv file class_data.csv and make it readble
df = pd.read_csv('class_data.csv')
```

```python
df_insta_followers = df[df['instagram_followers'] < 500] # Only instagram_followers below 500
```

```python
df_insta_followers.name = 'Total Insta followers' #Labeling the variable for plotting
instagram_followers = df_insta_followers.dropna() #Selecting the instagram followers column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(instagram_followers["instagram_followers"], kde=False) # Distploting with in KDE False otherwise it will start by minus numbers
plt.title('Insta followers') # Giving a title
plt.xlabel('Number of Instagram followers') # Giving a xlabel name
plt.show()
```

```python
mean_followers = df['instagram_followers'].mean() # Counting the mean
median_followers = df['instagram_followers'].median() # Counting the median

print(f'The mean Instagram followers are: {mean_followers} ')
print(f'The median Instagram followers are: {median_followers}')
```

```python

sns.distplot(instagram_followers['instagram_followers'], kde=False) #I need to drop the NA's, otherwise Seaborn won't plot
plt.title('Mean and Median of followers')
plt.axvline(mean_followers, 0, 100, color='red', label='mean') 
plt.axvline(median_followers, 0, 100, color='yellow', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
sns.boxplot(df['phones_owned']) # Boxplotting by phone owned
plt.title('Phone owned by users') # Giving a title
plt.xlabel('Amount') # Giving a xlabel name
plt.show() 
```

```python
df_apps_phone = df[df['apps_phone'] > 0] #Subsetting the dataframe to include only apps on the phone
sns.violinplot(df_apps_phone['apps_phone']) # Plot a violin
sns.swarmplot(df_apps_phone['apps_phone'], color='purple') # Plot swarmplot with in color purple
plt.title('Apps installed by users on their phone') # Giving a title
plt.ylabel('Frequency') # Giving a ylabel name
plt.xlabel('Amount') # Giving a xlabel name
plt.show() 
```
