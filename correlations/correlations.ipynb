```python
# Import modules
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv('funda-amsterdam.csv')
```

```python
df.head()
```

```python
df_subset = df[['area', 'price', 'year_built']]
sns.pairplot(df_subset);
```

```python
df = df[(df['price'] > 30000) & (df['area'] > 10)]
df_two_variables = df[['price', 'area']]
sns.lmplot(x='price', y='area', fit_reg=False, data=df_two_variables)
plt.show()
```

```python
#There is a big linearity relation as you can see, ofcourse there is some outliers. But in general it shows a linearity relation.
```
